```python
#DRAFT, MISTAKE
#Some commonly used correlation filtering methods have a tendency to drop more features than required. 
#This problem is amplified as datasets become larger and with more pairwise correlations above a specified threshold.
#If we drop more variables than necessary, less information will be available potentially leading to suboptimal model performance. 

#1 instead of 0.9, lower triangle instead of upper, 0.4 instead of 0.6

def corr_notok(X_train_cont, cut = 1) :
        
    # Get correlation matrix and lower triagle
    corr_matrix = X_train_cont.corr().abs()
    average_corr = corr_matrix.mean(axis = 1)
    lower_matrix = corr_matrix.where(np.tril(np.ones(corr_matrix.shape), k=1).astype(bool))
    
    drop = list()

        
    # For loop implements this pseudocode
    # For every cell in the lower triangle: 
        # If cell.value > 0.4: 
                # If mean(row_correlation) > mean(column_correlation):drop(column)
# Else: drop(row)
    for row in range(len(lower_matrix)-1):
        col_index = row + 1
        for col in range (col_index, len(lower_matrix)):
            if(corr_matrix.iloc[row, col] > cut):
                if(average_corr.iloc[row] > average_corr.iloc[col]): 
                    drop.append(row)
                else: 
                    drop.append(col)
    
    drop_set = list(set(drop))
    # dropcols_idx = drop_set
    dropcols_featurename = list(X_train_cont.columns[[item for item in drop_set]])
    
    return(dropcols_featurename)
```

```python
#Right code because of the sequential nature of the Python functions. Every decision to drop variables happens in order and is final.

def corr_ok(X_train_cont, cut = 1) :
       
    # Get correlation matrix and lower triagle
    corr_matrix = X_train_cont.corr().abs()
    average_corr = corr_matrix.mean(axis = 1)
    lower_matrix = corr_matrix.where(np.tril(np.ones(corr_matrix.shape), k=1).astype(bool))
    
    dropcols = list()
    
    new_table = pd.DataFrame(columns=(['var1', 'var2', 'v1.target', 
                                 'v2.target','corr_coef', 'drop' ]))
    
    for row in range(len(lower_matrix)-1):
        col_index = row + 1
        for col in range (col_index, len(lower_matrix)):
            if(corr_matrix.iloc[row, col] > cut):
                if(average_corr.iloc[row] > average_corr.iloc[col]): 
                    dropcols.append(row)
                    drop = corr_matrix.columns[row]
                else: 
                    dropcols.append(col)
                    drop = corr_matrix.columns[col]
                
                a = pd.Series([corr_matrix.index[row],
                lower_matrix.columns[col],
                average_corr[row],
                average_corr[col],
                lower_matrix.iloc[row,col],
                drop],
                index = new_table.columns)
        
                new_table = new_table.append(a, ignore_index = True)
    
    dropcols_featurename = calcDrop(new_table)
    
    return(dropcols_featurename)
```

```python
def calcDrop(new_table):
    # All variables with correlation > cutoff
    all_corr_vars = list(set(new_table['var1'].tolist() + new_table['var2'].tolist()))
    
    # All unique variables in drop column
    poss_drop = list(set(new_table['drop'].tolist()))

    # Keep any variable not in drop column
    keep = list(set(all_corr_vars).difference(set(poss_drop)))
     
    # Drop any variables in same row as a keep variable
    p = new_table[new_table['var1'].isin(keep)  | new_table['var2'].isin(keep) ][['var1', 'var2']]
    q = list(set(p['var1'].tolist() + p['var2'].tolist()))
    drop = (list(set(q).difference(set(keep))))

    # Remove drop variables from possible drop 
    poss_drop = list(set(poss_drop).difference(set(drop)))
    
    # subset new_table dataframe to include possible drop pairs
    m = new_table[new_table['var1'].isin(poss_drop)  | new_table['var2'].isin(poss_drop) ][['var1', 'var2','drop']]
        
    # remove rows that are decided (drop), take set and add to drops
    moredrops = set(list(m[~m['var1'].isin(drop) & ~m['var2'].isin(drop)]['drop']))
    for item in moredrops:
        drop.append(item)
         
    return drop
```
